```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.to_csv('practice_iris')
```

```python
df_color = pd.DataFrame({'color':['green','blue','yalow','green','blue'],'size':['small','large','medium','large','small']})
df_color
```

```python
dfm = pd.merge(df,df_color, how = 'inner', left_index = True, right_index = True)
dfm
```

```python
dfm['size_label'] = LabelEncoder().fit_transform(dfm['size'].values)
dfm
```

```python
dfmg = pd.get_dummies(dfm[['color']])
pf = pd.concat([dfm,dfmg], axis = 1)
pf
```

```python
dfmd = pd.get_dummies(dfm['color'])
af = pd.concat([dfm,dfmd], axis = 1)
af

#少一個中括號，欄位名稱會少「color_」
```
