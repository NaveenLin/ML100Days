```python
import numpy as np
import pandas as pd
```

```python
classes = ['1A', '1A', '1A', '1B', '1B', '1C', '1C', '1C']
name = ['Peter', 'John', 'Keven', 'JoJo', 'Mary', 'Eva', 'Tom', 'Simon']
math = [95, 95, 76, 90, 96, 84, 84, 98]
english = [64,88,85,88,78,60,63, 84]

tmpdic = {'class':classes, 'name':name, 'math':math, 'english':english}

students = pd.DataFrame(tmpdic)
students
```

```python
group = students.groupby('class')
group
```

```python
g1 = group.get_group('1A')
g1
```

```python
sex = ['boy','boy','girl','girl','boy','girl','girl', 'girl']
sexdic = {'sex':sex}
Sex = pd.DataFrame(sexdic)
Sex
```

```python
SS = pd.merge(students, Sex, left_index = True, right_index = True)
SS
```
