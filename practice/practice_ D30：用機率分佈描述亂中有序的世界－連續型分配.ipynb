```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import statistics
import seaborn as sns
from scipy import stats
```

```python
x = np.linspace(0,20,11)
x
```

```python
low = 2
high = 10
y = stats.uniform.pdf(x, low, high-low)
plt.plot(x, y, 'g--', linewidth = 4)
plt.grid(True)
plt.title('pdf of uniform(2,10)')
plt.show()
```

```python
cumsum_probs = stats.uniform.cdf(x, low, high-low)
plt.plot(x, cumsum_probs, 'y-', linewidth = 2)
plt.ylabel('P<=x')
plt.title('cdf of uniform(2,10)')
plt.xlabel('x')
plt.grid(True)
plt.show()
print(cumsum_probs)
```

```python
k = stats.uniform.ppf(cumsum_probs, low, high)
print(k)
```

```python
X = stats.uniform.rvs(low, high-low, size = 100)
plt.hist(X, bins= 13)
plt.title('random variable of uniform(size = 100)')
plt.show()
```

```python
stat_uniform = stats.uniform.stats(low, high- low, moments = 'mvks')
print(stat_uniform)
```
