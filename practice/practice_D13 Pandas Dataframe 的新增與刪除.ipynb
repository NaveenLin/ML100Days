```python
df1 = pd.DataFrame({'id':['A','B'],'Name': ['Bob','Andy']})
df1
```

```python
df2 = pd.DataFrame({'id':['C','D'], 'Name' : ['Evanna','Ruby']})
df2
```

```python
df12 = pd.concat([df1,df2]).reset_index(drop = True)
df12
```

```python
df3 = pd.DataFrame({'Name': ['Bob', 'Andy', 'Evanna','Ruby'], 'Age':[ 25,21,28,26]})
df3
```

```python
df4 = pd.merge(df12,df3, on = 'Name' )
df4
```

```python
df5 = pd.DataFrame({'hight':[190,175,160,162]})
df5
```

```python
df6 = df12.join(df5)
df6
```

```python
df7 = pd.DataFrame({'hight':[190,175]})
df8 = df12.join(df7)
df8
```

```python
df9 = pd.DataFrame({'Name': ['Bob', 'Andy', 'Evanna','Kyle','Anson'], 'Age':[ 25,21,28,26,35]})
df10 = pd.merge(df12, df9, how = 'inner', on = 'Name')
df10
```

```python
df11 = pd.DataFrame({'Name': ['Bob', 'Andy', 'Evanna','Kyle','Anson'], 'Age':[ 25,21,28,26,35]})
df12 = pd.merge(df12, df11, how = 'outer', on = 'Name')
df12
```

```python
df13 = pd.DataFrame({'Name': ['Bob', 'Andy', 'Evanna','Kyle','Anson'], 'Age':[ 25,21,28,26,35]})
df14 = pd.merge(df12, df13, how = 'right', on = 'Name')
df14
```

```python
df15 = pd.DataFrame({'Name': ['Bob', 'Andy', 'Evanna','Kyle','Anson'], 'Age':[ 25,21,28,26,35]})
df16 = pd.merge(df12, df15, how = 'left', on = 'Name')
df16
```
