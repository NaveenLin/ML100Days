```python
low = 1
high = 11
r = np.arange(low, high)
r
```

```python
r = np.arange(stats.randint.ppf(0.01, low, high),
              stats.randint.ppf(0.99, low, high),1)
r
```

```python
probs = stats.randint.pmf(r, low, high)
probs
```

```python
r
```

```python
plt.bar(r, probs)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('pmf DU(1,10)')
plt.show()
```

```python
cum_sum = stats.randint.cdf(r, low, high)
plt.plot(r, cum_sum )
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('cdf DU(1,10)')
plt.show()
```

```python
r = np.arange(0,2)
p = 0.4
probs  = stats.bernoulli.pmf(r, p)
plt.bar(r, probs)
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('pmf bernoulli(P=0.4)')
plt.show()
```

```python
cum_sum = stats.bernoulli.cdf(r,p)
plt.plot(r, cum_sum)
plt.xlabel('x')
plt.ylabel('P(X<= x)')
plt.title('cdf bernoulli(P=0.4)')
plt.show()
```
